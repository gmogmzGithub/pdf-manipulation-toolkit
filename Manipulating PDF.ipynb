```python
# Open the Source file
f = open('Harvard_Business_School.pdf', 'rb')
read_pdf = PdfReader(f)

# Create a writer object
write_pdf = PdfWriter()

# Read a page and rotate it
rotated_page = read_pdf.pages[4].rotate(90)

# Add the rotated page to the writer object
write_pdf.add_page(rotated_page)

# Save the writer object somewhere!
pdf_output = open('NewlyPdfHarvard_rotated.pdf', 'wb')
write_pdf.write(pdf_output)
pdf_output.close() # Close the new file
f.close() # Close the new file
```

```python
f = open('Harvard_Business_School.pdf', 'rb')
read_pdf = PdfReader(f)

pdf_text_all = [] # Create an empty list to hold the data
num_pages = len(read_pdf.pages)

for page in range(num_pages):
    one_page_text = read_pdf.pages[page].extract_text()
    pdf_text_all.append(one_page_text)

len(pdf_text_all) 
```

```python
with open('watermark_conf_2.pdf', 'rb') as watermark_file:
    read_watermark = PdfReader(watermark_file)
    watermark_page = read_watermark.pages[0]

with open('Harvard_Business_School.pdf', 'rb') as input_file:
    read_pdf = PdfReader(input_file)
    
    write_pdf = PdfWriter()

    for page in range(len(read_pdf.pages)):
        single_page = read_pdf.pages[page]
        single_page.merge_page(watermark_page)
        write_pdf.add_page(single_page)

    with open('Harvard_watermarked.pdf', 'wb') as output_file:
        write_pdf.write(output_file)
```
